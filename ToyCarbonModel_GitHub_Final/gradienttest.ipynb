```python
## Define cost and gradient functions
def fourDEnVar_cost_f(Yb_dash, hxbar, y, R_inv, w):

    bgrnd_term=np.dot(w.T,w)
    work1=np.dot(Yb_dash,w)+hxbar-y
    work2=np.dot(R_inv,work1)
    obs_term=np.dot(work1.T,work2)
    J=0.5*bgrnd_term+0.5*obs_term
                                
    return J

def fourDEnVar_cost_df(Yb_dash, hxbar, y, R_inv, w):

    work1=np.dot(Yb_dash,w)+hxbar-y
    work2=np.dot(R_inv, work1)
    df=w+np.dot(Yb_dash.T,work2)
      
    return df
```

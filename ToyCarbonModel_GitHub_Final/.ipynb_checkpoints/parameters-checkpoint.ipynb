```python
# Selecting a prior
import numpy as np
import matplotlib.pyplot as plt

F0=1
s0=0.01

p1min=0.5
p1max=5
p2min=0.5
p2max=5
k1min=0.03
k1max=0.9
k2min=0.01
k2max=0.12
p1range=np.linspace(p1min,p1max,100)
p2range=np.linspace(p2min,p2max,100)
k1range=np.linspace(k1min,k1max,100)
k2range=np.linspace(k2min,k2max,100)

p1=1
p2=1
k1=0.2
k2=0.1

accept_params=[]

for i in range(len(p1range)):
    for j in range(len(p2range)):
        for m in range(len(k1range)):
            for n in range(len(k2range)):
                
                c0=p1range[i]*p2range[j]*k2range[n]*s0/(k1range[m]**2)
                c1=((p1range[i]*k1range[m]+p2range[j]*k2range[n])*s0-p1range[i]*p2range[j]*k1range[m]*k2range[n])/(k1range[m]**2)
                c2=(F0-p1range[i]*k1range[m]-p2range[j]*k2range[n]+s0)/k1range[m]
                c3=-1

                coeffs=[c3,c2,c1,c0]
                cubicroots=np.roots(coeffs)
                #print(cubicroots)
                realroots = cubicroots.real[abs(cubicroots.imag)<1e-5]
                #print(realroots)
                if len(realroots)==1:
                    accept_params.append([p1range[i],p2range[j],k1range[m],k2range[n]])

np.shape(accept_params)                    

#x=np.linspace(-1,4,100)
#fx=[]
#for i in range(len(x)):
#    fx.append(c3*x[i]**3+c2*x[i]**2+c1*x[i]+c0)
#plt.plot(x,fx)
```

```python
fig = plt.figure().suptitle('Scatterplots')
plt.subplot(3,2,1)
plt.scatter(accept_params[:,0],accept_params[:,1])
plt.title('p1 vs p2')
plt.subplot(3,2,2)
plt.scatter(accept_params[:,0],accept_params[:,2])
plt.title('p1 vs k1')
plt.subplot(3,2,3)
plt.scatter(accept_params[:,0],accept_params[:,3])
plt.title('p1 vs k2')
plt.subplot(3,2,4)
plt.scatter(accept_params[:,1],accept_params[:,2])
plt.title('p2 vs k1')
plt.subplot(3,2,5)
plt.scatter(accept_params[:,1],accept_params[:,3])
plt.title('p2 vs k2')
plt.subplot(3,2,6)
plt.scatter(accept_params[:,2],accept_params[:,2])
plt.title('k1 vs k2')
```

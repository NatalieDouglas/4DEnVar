```python
import numpy as np
import matplotlib.pyplot as plt
import functools

def rk4(Xold, deltat, f):
    k1 = f(Xold)
    k2 = f(Xold + 1 / 2.0 * deltat * k1)
    k3 = f(Xold + 1 / 2.0 * deltat * k2)
    k4 = f(Xold + deltat * k3)
    delta = deltat * (k1 + 2 * k2 + 2 * k3 + k4) / 6.0
    return delta

def f(x, param=None):
    
    if np.all(param)==None:
        F = 1
        q1 = 1
        q2 = 1
        k1 = 0.2
        k2 = 0.1
        s1 = 0.01
    else:
        F, q1, q2, k1, k2, s1 = param
     
    # Initialize
    k = np.empty_like(x)
    k.fill(np.nan)
    k[0] = F*x[0]*x[1]/((x[0]+q1)*(x[1]+q2))+s1-k1*x[0]
    k[1] = k1*x[0]-k2*x[1]
    
    return k

def modelrun(x0, tf, deltat, discard, param):
    nx = len(x0)
    nt = int(tf / deltat) + 1
    xf = np.empty((nx, nt), order="F")
    xf.fill(np.nan)

    model = functools.partial(f, param=param)

    for time in range(discard):
        x0[:] = x0[:] + rk4(x0[:], deltat, model)

    xf[:, 0] = x0[:]

    for time in range(nt - 1):
        xf[:, time + 1] = xf[:, time] + rk4(xf[:, time], deltat, model)
    return xf

x0 = [1,1] # True initial conditions 
paramtrue = [1,1,1,0.2,0.1,0.01] # True parameters
t0=0 # Start time
tf = 1000 # Final time
deltat = 0.1 # Time step
discard = 0 # How many spin up integrations to ignore
t = np.arange(t0 + discard * deltat, tf + deltat / 2, deltat) # 

xtrue_traj = modelrun(x0, tf, deltat, discard, paramtrue)

#Plots
plt.figure().suptitle('Simple carbon system - Truth')
for i in range(2):
    plt.subplot(2,1,i+1)
    plt.plot(t,xtrue_traj[i],'k')
    plt.ylabel('x_'+str(i+1))
    plt.xlabel('time')
    plt.grid(True)
    plt.subplots_adjust(hspace=0.3)

plt.figure()
plt.plot(xtrue_traj[0],xtrue_traj[1])
```

```python
# get B matrix using Canadian Quick method
def getBcanadian(xt, diff_period, max_var, sample_size):
    _, total_steps = xt.shape
    assert total_steps >= sample_size, \
        f'model trajectory length {total_steps} must >= sample_size {sample_size}'

    sam_period = 1
    ind_sample_0 = np.arange(0, sample_size - diff_period, sam_period)
    sam_size = len(ind_sample_0)
    ind_sample_plus = ind_sample_0 + diff_period
    x_sample = xt[:, ind_sample_0] - xt[:, ind_sample_plus]

    Bcorr = np.corrcoef(x_sample)
    B = np.cov(x_sample)

    if max_var:
        alpha = max_var / np.amax(np.diag(B))
        B = alpha * B
    return B, Bcorr

lag         = int(0.1/deltat) # the lag, T, in number of time steps,
                    # 6 hours is about 0.05 time units, one period of system is about 1.8 time units
sample_size = 10000 # number of samples to estimate B (create a climatological matrix)
max_var     = None #2# # if set scales the estimated B to have a maximum variance
xB  = modelrun(xtrue_traj[:,0], sample_size*deltat, deltat, 0, paramtrue)
B,Bcorr     = getBcanadian (xB, lag, max_var, sample_size)

# set plot properties and then plot
mycmap      = 'BrBG'
vs          = [-np.amax(np.diag(B)),np.amax(np.diag(B))]
N1, N2 = np.shape(B)
plt.figure()
plt.pcolor(np.array(B).T, edgecolors="k", cmap=mycmap, vmin=vs[0], vmax=vs[1])
ymin, ymax = plt.ylim()
plt.ylim(ymax, ymin)
plt.colorbar(extend="both")
plt.title("matrix")
plt.xlabel("variable number")
plt.ylabel("variable number")
plt.xticks(np.arange(0.5, N1 + 0.5), np.arange(N1))
plt.yticks(np.arange(0.5, N2 + 0.5), np.arange(N2))
```

```python
# get simple B matrix
def getBsimple(xt, samfreq=2):
    err2 = 2
    _, total_steps = xt.shape
    # Precreate the matrix
    ind_sample = range(0, total_steps, samfreq)
    x_sample = xt[:, ind_sample]
    Bcorr = np.corrcoef(x_sample)

    B = np.cov(x_sample)
    alpha = err2/np.amax(np.diag(B))
    B = alpha*B
    return B, Bcorr

samplefreq=2
B_simp,Bcorr_simp = getBsimple(xB, samplefreq)

# set plot properties and then plot
mycmap      = 'BrBG'
vs          = [-np.amax(np.diag(B_simp)),np.amax(np.diag(B_simp))]
N1, N2 = np.shape(B_simp)
plt.figure()
plt.pcolor(np.array(B_simp).T, edgecolors="k", cmap=mycmap, vmin=vs[0], vmax=vs[1])
ymin, ymax = plt.ylim()
plt.ylim(ymax, ymin)
plt.colorbar(extend="both")
plt.title("matrix")
plt.xlabel("variable number")
plt.ylabel("variable number")
plt.xticks(np.arange(0.5, N1 + 0.5), np.arange(N1))
plt.yticks(np.arange(0.5, N2 + 0.5), np.arange(N2))

```

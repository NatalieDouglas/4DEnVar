```python
# Run the model for the truth trajectory - LORENZ96

def rk4(Xold, deltat, f):
    k1 = f(Xold)
    k2 = f(Xold + 1 / 2.0 * deltat * k1)
    k3 = f(Xold + 1 / 2.0 * deltat * k2)
    k4 = f(Xold + deltat * k3)
    delta = deltat * (k1 + 2 * k2 + 2 * k3 + k4) / 6.0
    return delta

def shift(x, n):
    return np.roll(x, -n)

def f(x, F):
    k = (shift(x, 1) - shift(x, -2)) * shift(x, -1) - x + F
    return k

n      = 12             # No. of state variables (No. of nodes in Lorenz-96, often n=40)
F      = 8.0            # Forcing F>5.0 guarantees chaos
deltat = 0.025          # Model timestep
tf     = 10.0            # The final time of the simulation
discard = 150

# Set the true initial conditions to something plus a perturbation
# This can be spun-up by discarding the first number of steps
x0     = np.repeat(10.0, n)
x0[1] += 0.05

# Set the timesteps
t = np.arange(0.0, tf+deltat/2.0, deltat)
nx = len(x0) # Size of state space
nt = int(tf / deltat) + 1 # Number of time steps (+1 as t=0 included also)
xf = np.empty((nx, nt), order="F") # Set-up the output matrix
xf.fill(np.nan)

# giving model parameters
model = functools.partial(f, F)
# Run a number of timesteps to spin-up the model (these are discarded)
for time in range(discard):
    x0[:] = x0[:] + rk4(x0[:], deltat, model)

# Initial conditions for part of integration which is kept
xf[:, 0] = x0[:]

# Timesteps solved via RK4
for time in range(nt - 1):
    xf[:, time + 1] = xf[:, time] + rk4(xf[:, time], deltat, model)

# Plot the truth
plotL96 (t, xtrue_traj, n, 'Lorenz 96 system (truth)')
plt.figure()
largest = max([np.abs(np.min(xt)), np.abs(np.max(xt))])
levs = np.linspace(-largest,largest,21)
mycmap = plt.get_cmap("BrBG", 21)
C = plt.contourf(np.arange(n), t, xt.T, cmap=mycmap, levels=levs)
plt.contour(np.arange(nx), t, xt.T, 10, colors="k", linestyles="solid")
plt.xlabel("variable number")
plt.ylabel("time")
plt.title("Hovmoller diagram: " + title)
plt.colorbar(C, extend="both")
```

```python
# Selecting a prior
import numpy as np
import matplotlib.pyplot as plt

F0=1
s0=0.01

p1min=0.5
p1max=5
p2min=0.5
p2max=5
k1min=0.03
k1max=0.9
k2min=0.01
k2max=0.12
nsteps=10
p1range=np.linspace(p1min,p1max,nsteps)
p2range=np.linspace(p2min,p2max,nsteps)
k1range=np.linspace(k1min,k1max,nsteps)
k2range=np.linspace(k2min,k2max,nsteps)

p1=1
p2=1
k1=0.2
k2=0.1

all_params=[]
params_1root=[]
params_3root_1pos=[]
params_3root_allpos=[]
counter = 0
counter_1root = 0
counter_1root_pos = 0
counter_2roots = 0
counter_3roots = 0
counter_3roots_allpos = 0
counter_3roots_nopos = 0
counter_3roots_1pos = 0
counter_3roots_2pos = 0
for i in range(len(p1range)):
    for j in range(len(p2range)):
        for m in range(len(k1range)):
            for n in range(len(k2range)):
                
                all_params.append([p1range[i],p2range[j],k1range[m],k2range[n]])
                
                c0=p1range[i]*p2range[j]*k2range[n]*s0/(k1range[m]**2)
                c1=((p1range[i]*k1range[m]+p2range[j]*k2range[n])*s0-p1range[i]*p2range[j]*k1range[m]*k2range[n])/(k1range[m]**2)
                c2=(F0-p1range[i]*k1range[m]-p2range[j]*k2range[n]+s0)/k1range[m]
                c3=-1

                coeffs=[c3,c2,c1,c0]
                cubicroots=np.roots(coeffs)
                #print(cubicroots)
                realroots = np.sort(cubicroots.real[abs(cubicroots.imag)<1e-5])
                #print(realroots)
                if len(realroots)==1:
                    counter_1root +=1
                    if realroots[0] > 0:
                        counter_1root_pos +=1
                        params_1root.append([p1range[i],p2range[j],k1range[m],k2range[n]])
                elif len(realroots)==2:
                    counter_2roots +=1
                elif len(realroots)==3:
                    counter_3roots +=1
                    if realroots[0] > 0 and realroots[1] > 0 and realroots[2] > 0:
                        counter_3roots_allpos +=1
                        params_3root_allpos.append([p1range[i],p2range[j],k1range[m],k2range[n]])
                    elif realroots[0] < 0 and realroots[1] > 0 and realroots[2] > 0:
                        counter_3roots_2pos +=1
                    elif realroots[0] < 0 and realroots[1] < 0 and realroots[2] > 0:
                        counter_3roots_1pos +=1
                        params_3root_1pos.append([p1range[i],p2range[j],k1range[m],k2range[n]])
                    else:
                        counter_3roots_nopos +=1
                counter += 1

print(np.shape(all_params))
print(counter, counter_1root, counter_2roots, counter_3roots)   
print(counter_1root_pos)
print(counter_3roots_nopos, counter_3roots_1pos, counter_3roots_2pos, counter_3roots_allpos )
#np.shape(accept_params)
#print(np.shape(accept_params)[0]/counter)

#x=np.linspace(-1,4,100)
#fx=[]
#for i in range(len(x)):
#    fx.append(c3*x[i]**3+c2*x[i]**2+c1*x[i]+c0)
#plt.plot(x,fx)
```

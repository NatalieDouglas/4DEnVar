```python
import numpy as np
import matplotlib.pyplot as plt
import functools
import importlib
import seaborn as sns
import fourDEnVar_engine
importlib.reload(fourDEnVar_engine)

def rk4(Xold, deltat, F, f):
    k1 = f(Xold, F)
    k2 = f(Xold + 1 / 2.0 * deltat * k1, F)
    k3 = f(Xold + 1 / 2.0 * deltat * k2, F)
    k4 = f(Xold + deltat * k3, F)
    delta = deltat * (k1 + 2 * k2 + 2 * k3 + k4) / 6.0
    return delta

def f(x, F, param=None):
    
    p2=1
    k1=0.2
    k2=0.1
    
    if param==None:
        p1 = 1
    else:
        p1 = param
     
    # Initialize
    k = np.empty_like(x)
    k.fill(np.nan)
    k[0] = F*x[0]*x[1]/((x[0]+p1)*(x[1]+p2))+0.01-k1*x[0]
    k[1] = k1*x[0]-k2*x[1]
    
    return k

def modelrun(x0, tf, deltat, F, param):
    nx = len(x0)
    nt = int(tf / deltat) + 1
    xf = np.empty((nx, nt), order="F")
    xf.fill(np.nan)

    model = functools.partial(f, param=param)

    xf[:, 0] = x0[:]

    for time in range(nt - 1):
        xf[:, time + 1] = xf[:, time] + rk4(xf[:, time], deltat, F[time], model)
    return xf

x0 = [1,1] # True initial conditions 
#paramtrue = [1.04,1.35,0.3,0.08] # True parameters: p1, p2, k1, k2
paramtrue = [1.04] # True 1
t0=0 # Start time
tf = 1000 # Final time
deltat = 1 # Time step
t = np.arange(t0, tf + deltat / 2, deltat) # 

## forcing calculation
#forcing=np.ones(len(t))
Tm = 10*deltat
a = np.exp(-deltat/Tm)
b = np.sqrt(1-a**2)
sigma = 0.5
w = np.random.normal(0, 1, len(t))
p0 = 1

m = np.zeros(len(t))
for i in range(0,len(t)-1):
    m[i+1] = a*m[i]+b*sigma*w[i]

forcing = p0*np.exp(m)
xtrue_traj = modelrun(x0, tf, deltat, forcing, paramtrue)

##Plots
#plt.figure().suptitle('Simple carbon system - Truth')
#for i in range(2):
#    plt.subplot(2,1,i+1)
#    plt.plot(t,xtrue_traj[i],'k')
#    plt.ylabel('x_'+str(i+1))
#    plt.xlabel('time')
#    plt.grid(True)
#    plt.subplots_adjust(hspace=0.3)

#plt.figure()
#plt.plot(xtrue_traj[0],xtrue_traj[1])

plt.figure(figsize=(10,4))
plt.plot(t,xtrue_traj[0],label='x1')
plt.plot(t,xtrue_traj[1],label='x2')
plt.plot(t,forcing,label='forcing')
plt.legend()
plt.xlabel('Time')
plt.ylabel('State/Forcing')
```

```python
# Get ensemble
nens = 50
prior_err = 0.25

def test_bnds(xbi, val_bnds):

    for xi in enumerate(xbi):
        if val_bnds[xi[0]][0] <= xi[1] <= val_bnds[xi[0]][1]:
            continue
        else:
            return False
    return True


def generate_z_ens(z_dic, size_ens=20, prior_err=0.25):

    z_keys = list(z_dic.keys())
    Xb = []
    val_bnds = []
    for zkey in z_keys:
        Xb.append(z_dic[zkey][0])
        val_bnds.append(z_dic[zkey][1])
    
    Xb = np.array(Xb)
    Xb_sd = Xb * prior_err
    B_mat = np.eye(len(Xb))*((Xb_sd)**2)

    ens = []
    i = 0
    #np.random.seed(seed_val)
    while len(ens) < size_ens:
        Xbi = np.random.multivariate_normal(Xb, B_mat)
        i += 1
        if test_bnds(Xbi, val_bnds) is True:
            ens.append(Xbi)
        else:
            continue
  
    return np.array(ens)

prior_values = {
    #'a_x1': [0, (-1, 1)],
    #'b_x2': [0, (-1, 1)],
    'c_p1': [2, (0, 4)],
    #'d_p2': [1, (0.5, 1.5)],
    #'e_k1': [0.2, (0.1, 0.3)],
    #'f_k2': [0.1, (0.01, 0.11)],
    #'c_p1': [2.75, (0.5, 5)],
    #'d_p2': [2.75, (0.5, 5)],
    #'e_k1': [0.465, (0.03, 0.9)],
    #'f_k2': [0.065, (0.01, 0.12)],
    }
    
z_ens = generate_z_ens(prior_values, size_ens=nens,
                prior_err=prior_err)
#print(z_ens)
z_mean=np.mean(z_ens)
z_ens=z_ens.T
print(z_mean)
print(np.shape(z_ens))   

paramnum=0
weights = np.ones_like(z_ens.T) / float(len(z_ens))
palette=sns.color_palette("colorblind", 11)
xlabels=['p1','p2','k1','k2']
fig = plt.figure().suptitle('Distributions')
sns.distplot(z_ens, kde=True, color=palette[6], hist_kws={'weights': weights}, label='prior')
plt.axvline(z_mean, color='k', linestyle='--') 
plt.tight_layout()
#plt.plot('Prior Value','k')
#plt.ylabel('x_'+str(i+1))
plt.xlabel(xlabels[paramnum])
#plt.grid(True)
#plt.subplots_adjust(hspace=0.3)
z_ens
```

```python
# Get observation vector and cov mat
obs_err=0.1
y=np.concatenate(xtrue_traj.T,axis=0)
R=obs_err*np.diag(y)

# get h(zbar)
#hzbar_mat = modelrun(z_mean[0:2], tf, deltat, forcing, z_mean[2:6])
hzbar_mat = modelrun(x0, tf, deltat, forcing, z_mean)
hzbar=np.concatenate(hzbar_mat.T,axis=0)
rmsediff_prior_1=(hzbar_mat[0]-xtrue_traj[0])**2.0
rmsemean_prior_1=np.mean(rmsediff_prior_1)
rmse_prior_1=rmsemean_prior_1**0.5
rmsediff_prior_2=(hzbar_mat[1]-xtrue_traj[1])**2.0
rmsemean_prior_2=np.mean(rmsediff_prior_2)
rmse_prior_2=rmsemean_prior_2**0.5

# get h(Z)
hZ=np.zeros((len(t)*2,nens))
hZ_arr=np.zeros((2,len(t),nens))
np.shape(hZ)
for i in range(0,nens):
    hZ_arr[:,:,i]=modelrun(x0, tf, deltat, forcing, z_ens[0,i])
    hZ[:,i]=np.concatenate(hZ_arr[:,:,i].T,axis=0)
    
# run 4DEnVar
DA=fourDEnVar_engine.fourDEnVar_engine(z_ens, hZ, y, R, hzbar)
xa=DA.xa
Xa=DA.Xa
anal=DA.xa_analytical
#Xanal=DA.Xa_analytical
print(xa)
print(anal)

hanal_mat = modelrun([1,1], tf, deltat, forcing, xa)
hanal = np.concatenate(hanal_mat.T,axis=0)
rmsediff_anal_1=(hanal_mat[0]-xtrue_traj[0])**2.0
rmsemean_anal_1=np.mean(rmsediff_anal_1)
rmse_anal_1=rmsemean_anal_1**0.5
rmsediff_anal_2=(hanal_mat[1]-xtrue_traj[1])**2.0
rmsemean_anal_2=np.mean(rmsediff_anal_2)
rmse_anal_2=rmsemean_anal_2**0.5
```

```python
import numpy as np
import matplotlib.pyplot as plt
import functools
import importlib
import seaborn as sns
import fourDEnVar_engine
importlib.reload(fourDEnVar_engine)

def rungekutta4(Xold, deltat, F, modelf):

    rk1 = modelf(Xold, F)
    rk2 = modelf(Xold + (1 / 2.0) * deltat * rk1, F)
    rk3 = modelf(Xold + (1 / 2.0) * deltat * rk2, F)
    rk4 = modelf(Xold + deltat * rk3, F)
    deltax = deltat * (rk1 + 2 * rk2 + 2 * rk3 + rk4) / 6.0
    return deltax


def f(params, x, Ft):
    
    if np.all(params)==None:
        p1 = 1
        p2 = 1
        k1 = 0.2
        k2 = 0.1
    else:
        p1, p2, k1, k2 = params
     
    # Initialize
    k = np.empty_like(x)
    k.fill(np.nan)
    k[0] = Ft*x[0]*x[1]/((x[0]+p1)*(x[1]+p2))+0.01-k1*x[0]
    k[1] = k1*x[0]-k2*x[1]
    
    return k

def modelrun(x0, tf, deltat, F, param):
    nx = len(x0)
    nt = int(tf / deltat) + 1
    xf = np.empty((nx, nt), order="F")
    xf.fill(np.nan)

    #model = functools.partial(f, param=param)
    model = functools.partial(f, param)

    xf[:, 0] = x0[:]

    for time in range(nt - 1):
        xf[:, time + 1] = xf[:, time] + rungekutta4(xf[:, time], deltat, F[time], model)
    return xf

x0 = [1,1] # True initial conditions 
#paramtrue = [1.04,1.35,0.23,0.08] # A1-A11 experiment
#paramtrue = [2.44,2.45,0.11,0.031] # B1-B6 experiment
#paramtrue = [2.44,2.45,0.11,0.011] # C1 experiment
#paramtrue = [0.77,2.73,0.033,0.025] # C2 experiment
#paramtrue = [1.14,1.55,0.23,0.11] # C3 experiment
paramtrue = [4.6,1.45,0.63,0.011] # C4 experiment

t0=0 # Start time
tf = 1000 # Final time
discard = 200
deltat = 1 # Time step
t = np.arange(t0, tf + discard + deltat / 2, deltat) # 

forcing = np.load('forcing.npy')
xtrue_traj = modelrun(x0, tf+discard, deltat, forcing, paramtrue)

#np.save('xtrue_parametersA1-A11.npy',xtrue_traj)
#np.save('xtrue_parametersB1-B6.npy',xtrue_traj)
#np.save('xtrue_parametersC1.npy',xtrue_traj)
#np.save('xtrue_parametersC2.npy',xtrue_traj)
#np.save('xtrue_parametersC3.npy',xtrue_traj)
np.save('xtrue_parametersC4.npy',xtrue_traj)

plt.figure(figsize=(10,4))
plt.plot(t-discard,xtrue_traj[0],label=r'$x_1$')
plt.plot(t-discard,xtrue_traj[1],label=r'$x_2$')
plt.plot(t-discard,forcing,label='forcing')
plt.axvline(0,color='k', linestyle='dashed')
plt.legend()
plt.xlabel('Time')
plt.ylabel('State/Forcing')
```
